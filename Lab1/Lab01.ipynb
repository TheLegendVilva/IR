```python
import re
from collections import defaultdict
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from nltk.stem import PorterStemmer

# Define additional articles to add to the scikit-learn stop words
additional_articles = {"a", "an", "the"}

# Create a set of custom stop words that includes both scikit-learn's stop words and the additional articles
custom_stop_words = ENGLISH_STOP_WORDS.union(additional_articles)

# Initialize a dictionary to store word frequencies
word_freq = defaultdict(int)

# Initialize the Porter Stemmer
stemmer = PorterStemmer()

# Open and read the text file
with open("output.txt", "r") as file:
    for line in file:
        # Tokenize the text using regular expressions (split by non-word characters)
        words = re.findall(r'\b\w+\b', line.lower())
        
        # Update word frequencies after removing stemming and custom stopwords
        for word in words:
            if word not in custom_stop_words:  # Check if the word is not a custom stop word
                original_word = stemmer.stem(word)  # Stem the word
                word_freq[original_word] += 1

# Display the word frequencies without stemming and custom stopwords
for word, freq in word_freq.items():
    print(f"{word}: {freq}")
```

```python
sorted_word_freq = dict(sorted(word_freq.items(), key=lambda item: item[1], reverse=True))
```

```python
import numpy as np
freq = np.array(list(sorted_word_freq.values()))
```

```python
rank = np.arange(1,freq.shape[0]+1)
```

```python
log_freq = np.log(freq).reshape(-1,1)
```

```python
log_rank = np.log(rank).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(log_rank, log_freq)
```

```python
c = lr.intercept_
```

```python
m = lr.coef_
```
